```python
import matplotlib.pyplot as plt
import pandas as pd
import asyncio
import seaborn as sns
%matplotlib widget
visualization = plt.figure()

async def refresh_visualization():
    for iteration in range(100):
        
        plt.clf()
        
        data_frame = pd.read_csv('output.csv')
        sns.scatterplot(data=data_frame, x='polarity', y='subjectivity', hue='prediction', style='title')
        
        plt.xlabel('Polarity')
        plt.ylabel('Subjectivity')
        plt.legend()

        visualization.canvas.draw()
        await asyncio.sleep(1)


event_loop = asyncio.get_event_loop()
event_loop.create_task(refresh_visualization())
```
